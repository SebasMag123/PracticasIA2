```python
#Importar pandas
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
datos =  pd.read_csv("P1-RegresionLineal.csv")
df = pd.DataFrame(datos)
plt.plot(df['equipos'],df['costo'],"b.")
plt.xlabel("Equipos")
plt.ylabel("Costo")
```

```python
#Convertir los datos a valores reales(1000,10000) 
df['equipos'] = df['equipos'] *1000
df['equipos'] = df['equipos'].astype('int')

df['costo']= df['costo']*10000
df['costo'] = df['costo'].astype('int')
df.head(10)
```

```python
#Graficar los datos reales
plt.plot(df['equipos'],df['costo'],"b.")
plt.xlabel("Equipos")
plt.ylabel("Costo")
```

```python
#General el modelo de regresion lineal
from sklearn.linear_model import LinearRegression
linReg = LinearRegression()
linReg.fit(df['equipos'].values.reshape(-1,1),df['costo'].values)
```

```python
#Mostrar theta
linReg.coef_
```

```python
#Graficar el modelo con el conjunto de datos
x_min_max=np.array([[df['equipos'].min()],[df['equipos'].max()]])
yTraningPred = linReg.predict(x_min_max)
```

```python
plt.plot(x_min_max,yTraningPred,"g--")
plt.plot(df['equipos'],df['costo'],'b.')
plt.xlabel("Equipos")
plt.ylabel("Costo")
plt.show()
```
